```python
# Imports for first TensorFlow example
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

## Model 1. Predict y-output for a given x-input

1. Build a Keras model, the simplest ml model, with one density (layer) and one unit (neuron)
2. Input random data (with a formula of 2x-1) through numpy
3. Fit a model with 500 epochs(generations)

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)

print(model.predict([10.0]))
```
