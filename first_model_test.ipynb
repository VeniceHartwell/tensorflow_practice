## Model 1. Predict y-output for a given x-input

1. Build a Keras model, the simplest ml model, with one density (layer) and one unit (neuron)
2. Input random data (with a formula of 2x-1) through numpy
3. Fit a model with 500 epochs(generations)

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)

print(model.predict([10.0]))
```

## Model 2. Image Recognition

```python
fashion_mnist = keras.datasets.fashion_mnist # import fashion images with 10 sections (index 9, what we're using, is shoes)
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # 28x28 is the size of the input images
    keras.layers.Dense(128, activation=tf.nn.relu), # this middle layer has 128 functions to adjust the output of the next layer
    keras.layers.Dense(10, activation=tf.nn.softmax) # this layer is 10 units 'wide' representing the number of different items of clothing in the dataset
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
             loss='sparse_categorical_crossentropy')

model.fit(train_images, train_labels, epochs=5) # train the model

# test_loss, test_acc = model.evaluate(test_images, test_labels) # evaluate the model against the data which wasn't used (test data)

predictions = model.predict(test_images)
```

### Accuracy

```python
# visualize prediction data
predictions[0]
```

```python
def accuracy_check(predictions, test_labels):
    count=0
    for i in range(len(test_labels)): # loop through all test labels

        best_guess = 0.
        for c in range(len(predictions[i])): # save the best guess' label in each prediction array by using the highest value
            if predictions[i][c] > best_guess:
                best_guess = predictions[i][c]
                predicted_label = c

        if predicted_label != test_labels[i]: # check the accuracy of the model
            count+=1

        #print(f'Predictions guessed label {predicted_label} for the test image {test_labels[i]}')
    print(f'Final Score: {(1-(count/10000))*100}%\nThere were {count} incorrect matches out of {len(test_labels)}.')
```

```python
accuracy_check(predictions, test_labels)
```

### Revising the model for accuracy: more epochs

```python
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # 28x28 is the size of the input images
    keras.layers.Dense(128, activation=tf.nn.relu), # this middle layer has 128 functions to adjust the output of the next layer
    keras.layers.Dense(10, activation=tf.nn.softmax) # this layer is 10 units 'wide' representing the number of different items of clothing in the dataset
])

model2.compile(optimizer=tf.keras.optimizers.Adam(),
             loss='sparse_categorical_crossentropy')

model2.fit(train_images, train_labels, epochs=50) # train the model

predictions2 = model2.predict(test_images)
```

```python
accuracy_check(predictions2, test_labels)
```

### Revising the model for accuracy: more layers

```python
model3 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # 28x28 is the size of the input images
    keras.layers.Dense(128, activation=tf.nn.relu), # this middle layer has 128 functions to adjust the output of the next layer
    keras.layers.Dense(128, activation=tf.nn.relu), # ADDED IN THIRD MODEL
    keras.layers.Dense(10, activation=tf.nn.softmax) # this layer is 10 units 'wide' representing the number of different items of clothing in the dataset
])

model3.compile(optimizer=tf.keras.optimizers.Adam(),
             loss='sparse_categorical_crossentropy')

model3.fit(train_images, train_labels, epochs=5) # train the model

predictions3 = model3.predict(test_images)
```

```python
accuracy_check(predictions3, test_labels)
```

### Revising the model for accuracy: a different optimizer

```python
model4 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # 28x28 is the size of the input images
    keras.layers.Dense(128, activation=tf.nn.relu), # this middle layer has 128 functions to adjust the output of the next layer
    keras.layers.Dense(10, activation=tf.nn.softmax) # this layer is 10 units 'wide' representing the number of different items of clothing in the dataset
])

model4.compile(optimizer=tf.keras.optimizers.SGD(),
             loss='sparse_categorical_crossentropy')

model4.fit(train_images, train_labels, epochs=5) # train the model

predictions4 = model4.predict(test_images)
```

```python
accuracy_check(predictions4, test_labels)
```

### Revising the model for accuracy: even more epochs + 2 layers

```python
model5 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # 28x28 is the size of the input images
    keras.layers.Dense(128, activation=tf.nn.relu), # this middle layer has 128 functions to adjust the output of the next layer
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax) # this layer is 10 units 'wide' representing the number of different items of clothing in the dataset
])

model5.compile(optimizer=tf.keras.optimizers.Adam(),
             loss='sparse_categorical_crossentropy')

model5.fit(train_images, train_labels, epochs=100) # train the model

predictions5 = model5.predict(test_images)
```

```python
accuracy_check(predictions5, test_labels)
```
